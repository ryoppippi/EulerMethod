## (1)

$$
\begin{eqnarray*}
y^2 &=& x^4(1-x^2) \\
y &=& \sqrt{x^4(1-x^2)}  \\
y &=& -\sqrt{x^4(1-x^2)} 
\end{eqnarray*}
$$


$$
F = y^2 - x^4(1-x^2) =0
$$
$$
\begin{eqnarray*}
\frac{dy}{dx} &=& -\frac{Fx(x,y)}{Fy(x,y)}  \\
   &=& -\frac{3x^5-2x^3}{y} \\
\end{eqnarray*}
$$
$$
\begin{eqnarray*}
dy=-\frac{3x^5-2x^3}{y_{n}}dx \\
y_{n+1} = y_{n} -\frac{3x^5-2x^3}{y_{n}}dx
\end{eqnarray*}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pda
import math
```

```python
def f1(x):
    return (x*x*x*x*(1-x*x))
def f2(x,y):
    return -((3*x*x-2)*x*x*x)/y
```

```python
dx = 0.00001
xmax = 10;
xBegin = -0.99999 #始まりのx座標を決めておく。定義域は-1<x<1より、適当に決める

#上の式より、xに対応するyは当然二箇所である。ここではまずyが正の点において調べる。
x = []
y = []
xPrev = xBegin
yPrev = math.sqrt(f1(xBegin)) #正のy座標を計算
while f1(xPrev) >= 0:
    yPrev = dx*f2(xPrev,yPrev) + yPrev
    x.append(xPrev)
    y.append(yPrev)
    xPrev += dx
plt.plot(x,y,"b")

#次にyが負の時
x = []
y = []
xPrev = xBegin
yPrev = - math.sqrt(f1(xBegin))#負のy座標を計算
while f1(xPrev) >= 0:
    yPrev = dx*f2(xPrev,yPrev) + yPrev
    x.append(xPrev)
    y.append(yPrev)
    xPrev += dx
    
plt.plot(x,y,"g")
plt.show()
```

## (2)

$$
\begin{eqnarray*}
y^2 = x^2(x+1) \\
y = \sqrt{x^2(x+1)}  \\
y = -\sqrt{x^2(x+1)} 
\end{eqnarray*}
$$



$$
F = y^2 - x^2(x+1) =0
$$
$$
\begin{eqnarray*}
\frac{dy}{dx} &=& -\frac{Fx(x,y)}{Fy(x,y)}  \\
   &=& -\frac{3x^2+2x}{2y} \\
\end{eqnarray*}
$$
$$
\begin{eqnarray*}
dy &=& -\frac{3x^2+2x}{2y}dx \\
y_{n+1} &=& y_{n} -\frac{3x^2+2x}{2y_{n}}dx
\end{eqnarray*}
$$

```python
def f3(x):
    return x*x*(x+1)
def f4(x,y):
    return -((3*x+2)*x)/2/y
```

```python
dx = 0.001
xEnd = 10;
xBegin = -0.99 #始まりのx座標を決めておく。定義域は-1=<xより、適当に決める

#上の式より、xに対応するyは当然二箇所である。ここではまずyが正の点において調べる。
x = []
y = []
xPrev = xBegin
yPrev = math.sqrt(f3(xBegin)) #正のy座標を計算
while xPrev <= xEnd:
    yPrev = dx*f4(xPrev,yPrev) + yPrev
    x.append(xPrev)
    y.append(yPrev)
    xPrev += dx
plt.plot(x,y,"b")

#次にyが負の時
x = []
y = []
xPrev = xBegin
yPrev = - math.sqrt(f3(xBegin))#負のy座標を計算
while xPrev <= xEnd:
    yPrev = dx*f4(xPrev,yPrev) + yPrev
    x.append(xPrev)
    y.append(yPrev)
    xPrev += dx
    
plt.plot(x,y,"g")
plt.show()
```

## (3)


$$
\begin{eqnarray*}
\frac{du}{dt} = u(1-u)
\end{eqnarray*}
$$
$$
\begin{eqnarray*}
u_{n+1} = u_{n} + u_{n}(1-u_{n})dt
\end{eqnarray*}
$$



```python
def f5(u):
    return u*(1-u)
```

この式ではカオスを再現するために、少々刻み幅を大きくする

```python
dt = 1
tEnd = 10
tBegin = -10 #始まりは適当に

#上の式より、xに対応するyは当然二箇所である。ここではまずyが正の点において調べる。
t = []
u = []
tPrev = tBegin
uPrev = 0.5 #初期条件
while tPrev <= tEnd:
    uPrev = dt*f5(uPrev) + uPrev
    t.append(tPrev)
    u.append(uPrev)
    tPrev += dt
plt.plot(t,u,"b")
plt.show
```

```python
### ではdtを試しに2.8と置いてみる
dt = 2.8
tEnd = 10;
tBegin = -10 #始まりは適当に

#上の式より、xに対応するyは当然二箇所である。ここではまずyが正の点において調べる。
t = []
u = []
tPrev = tBegin
uPrev = 1/2 #初期条件
while tPrev <= tEnd:
    uPrev = dt*f5(uPrev) + uPrev
    t.append(tPrev)
    u.append(uPrev)
    tPrev += dt
plt.plot(t,u,"b")
plt.show
```

これがカオスと呼ばれる現象である。
